```python
import Ipynb_importer
import matplotlib.pyplot as plt
import MUSIC
import numpy as np
from scipy import signal
```

```python
SNR = -25
snapshots = 200
sensorNum = 8
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
fig = plt.figure(0)
plt.subplot(2, 2, 1)
plt.semilogy(theta, abs(P)/np.max(abs(P)))
plt.xlabel('angle(degree)')
plt.ylabel('P(dB)')
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum))

SNR = -10
snapshots = 200
sensorNum = 8
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
plt.subplot(2, 2, 2)
plt.semilogy(theta, abs(P)/np.max(abs(P)))
plt.xlabel('angle(degree)')
plt.ylabel('P(dB)')
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum))

SNR = 10
snapshots = 200
sensorNum = 8
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
plt.subplot(2, 2, 3)
plt.semilogy(theta, abs(P)/np.max(abs(P)))
plt.xlabel('angle(degree)')
plt.ylabel('P(dB)')
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum))

SNR = 25
snapshots = 200
sensorNum = 8
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
plt.subplot(2, 2, 4)
plt.semilogy(theta, abs(P)/np.max(abs(P)))
plt.xlabel('angle(degree)')
plt.ylabel('P(dB)')
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum))

fig.set_size_inches(10, 10)
fig.savefig('Different SNR using MUSIC')
plt.show()
```
