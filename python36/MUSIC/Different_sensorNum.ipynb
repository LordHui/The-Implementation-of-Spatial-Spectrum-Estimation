```python
import Ipynb_importer
import matplotlib.pyplot as plt
import MUSIC
import numpy as np
from scipy import signal
```

```python
theta_S = np.array([-10, 0, 10])
snapshots = 200
SNR = 25

N = 1000

STD_RMES_list = []
SS_RMES_list = []
sensorNumVals = range(5, 31)
for sensorNum in sensorNumVals:
    RMES = 0
    
    for itr in range(N):
        theta, P_mu = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
        peakIdx = signal.argrelmax(abs(P_mu))[0]
        peakVals = abs(P_mu[peakIdx])
        idx = np.argsort(-peakVals)
        peakIdx = peakIdx[idx[: len(theta_S)]]
        theta_h = theta[peakIdx]    # theta hat            
        theta_h.sort()
        
        res = theta_S - theta_h
        RMES = RMES + np.sqrt(np.dot(res.T, res)/len(theta_S))
        
    RMES = RMES/N
    STD_RMES_list.append(RMES)
    
    RMES = 0
    
    for itr in range(N):
        theta, P_mu = MUSIC.SS_MUSIC(SNR, snapshots, sensorNum)
        peakIdx = signal.argrelmax(abs(P_mu))[0]
        peakVals = abs(P_mu[peakIdx])
        idx = np.argsort(-peakVals)
        peakIdx = peakIdx[idx[: len(theta_S)]]
        theta_h = theta[peakIdx]    # theta hat            
        theta_h.sort()
        
        res = theta_S - theta_h
        RMES = RMES + np.sqrt(np.dot(res.T, res)/len(theta_S))
        
    RMES = RMES/N
    SS_RMES_list.append(RMES)

STD_RMES_list = np.array(STD_RMES_list)
SS_RMES_list = np.array(SS_RMES_list)

font = {
    'family': 'monospace',
    'weight': 'normal',
    'size': 10.5,
}
fig = plt.figure(0)
plt.plot(sensorNumVals, STD_RMES_list, label='MUSIC', marker='o', linestyle='-', color='k', linewidth=4.0)
plt.hold
plt.plot(sensorNumVals, SS_RMES_list, label='SS-MUSIC', marker='*', linestyle='--', color='k', linewidth=4.0)
plt.tick_params(labelsize=10.5)
plt.grid(True, linestyle='--', color='k', linewidth=2.0)
plt.legend(prop=font)
plt.xlabel('The number of sensor', font)
plt.ylabel('RMES', font)
plt.title('The number of test: ' + str(N), font)
fig.set_size_inches(7, 7)
plt.show()
fig.savefig('Two mathods in different sensorNum')
```
