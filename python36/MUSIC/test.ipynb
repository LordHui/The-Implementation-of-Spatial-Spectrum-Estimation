```python
import Ipynb_importer
import matplotlib.pyplot as plt
import MUSIC
import numpy as np
from scipy import signal
```

```python
theta, P = MUSIC.STD_MUSIC(SNR=25)
fig = plt.figure(0)
plt.semilogy(theta, abs(P)/np.max(abs(P)))
plt.xlabel('angle(degree)')
plt.ylabel('P(dB)')
fig.set_size_inches(10, 10)
fig.savefig('MUSIC with coherent signals')
plt.show()
```
