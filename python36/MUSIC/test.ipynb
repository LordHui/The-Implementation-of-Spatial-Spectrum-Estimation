```python
import Ipynb_importer
import matplotlib.pyplot as plt
import MUSIC
import numpy as np
from scipy import signal
```

```python
SNR = 25
snapshots = 200
sensorNum = 5
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
fig = plt.figure(0)
font = {
    'family': 'monospace',
    'weight': 'normal',
    'size': 10.5,
}

plt.subplot(2, 2, 1)
plt.semilogy(theta, abs(P)/np.max(abs(P)), 'k-', linewidth=4.0)
plt.grid(True, linestyle='--', color='k', linewidth=2.0)
plt.xlabel('angle(degree)', font)
plt.ylabel('P(dB)', font)
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum), font)
plt.tick_params(labelsize=10.5) 

SNR = 25
snapshots = 200
sensorNum = 8
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
plt.subplot(2, 2, 2)
plt.semilogy(theta, abs(P)/np.max(abs(P)), 'k-', linewidth=4.0)
plt.grid(True, linestyle='--', color='k', linewidth=2.0)
plt.xlabel('angle(degree)', font)
plt.ylabel('P(dB)', font)
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum), font)
plt.tick_params(labelsize=10.5) 

SNR = 25
snapshots = 200
sensorNum = 16
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
plt.subplot(2, 2, 3)
plt.semilogy(theta, abs(P)/np.max(abs(P)), 'k-', linewidth=4.0)
plt.grid(True, linestyle='--', color='k', linewidth=2.0)
plt.xlabel('angle(degree)', font)
plt.ylabel('P(dB)', font)
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum), font)
plt.tick_params(labelsize=10.5) 

SNR = 25
snapshots = 200
sensorNum = 24
theta, P = MUSIC.STD_MUSIC(SNR, snapshots, sensorNum)
plt.subplot(2, 2, 4)
plt.semilogy(theta, abs(P)/np.max(abs(P)), 'k-', linewidth=4.0)
plt.grid(True, linestyle='--', color='k', linewidth=2.0)
plt.xlabel('angle(degree)', font)
plt.ylabel('P(dB)', font)
plt.title('SNR :' + str(SNR) + ' snapshots :' + str(snapshots) + ' sensorNum :' + str(sensorNum), font)
plt.tick_params(labelsize=10.5) 

fig.set_size_inches(10, 10)
fig.savefig('Different sensorNum using MUSIC')
plt.show()
```
